```python
import pandas as pd
import numpy as np
import names
import random 
from faker import Faker
```

```python
Description = ["Business","Investing", "Education", "Personal Use", "Traveling", "Medical_Expenses", "Weddings", "Emergence_Expenses"]
Employment = ["Employed","UnEmployed"]
```

```python
sheet = pd.DataFrame(columns=["ID","Date","Name","Age","Income","Employment_status", "Purpose","Loan_Amount","Credit_History"])
sheet["ID"] = range(1,1001)
sheet["Date"] = pd.to_datetime(np.random.choice(pd.date_range("2010-01-01", "2024-12-31"),size=1000))
sheet["Name"] = [names.get_full_name() for _ in range(1000)]
sheet["Age"] = np.random.randint(21,100, size= 1000)
sheet["Income"] = np.random.randint(2000,100000, size=1000)
unemployment_rate = 0.2
sheet["Employment_status"] = np.random.choice(Employment, size=1000, p=[1 - unemployment_rate, unemployment_rate])
sheet["Purpose"] = np.random.choice(Description, size = 1000)
sheet["Loan_Amount"] = np.random.randint(501,1000000, size=1000)

sheet.loc[sheet["Employment_status"] == "UnEmployed", "Income"] = 0

conditions = [
    sheet["Employment_status"] == "UnEmployed",   # Unemployed → Poor
    sheet["Income"] <= 10000,                     # Income ≤ 10,000 → Average
    sheet["Income"] > 10000                       # Income > 10,000 → Good
]

choices = ["Poor", "Average", "Good"]

sheet["Credit_History"] = np.select(conditions, choices, default="Average")
```

```python




sheet.to_csv("creditrisk.csv", index=False)
```
